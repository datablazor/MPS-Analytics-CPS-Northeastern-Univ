```python
factors = ['Cylinders','Displacement','Horsepower', 'Weight', 'Acceleration', 'Model Year','US Made']
X = pd.DataFrame(df[factors].copy())
y = df['MPG'].copy()

X = X.apply(pd.to_numeric, errors='coerce')
y = y.apply(pd.to_numeric, errors='coerce')
```

```python
X.fillna(0, inplace=True)
y.fillna(0, inplace=True)
```

```python
y.mean()
```

```python
X = StandardScaler().fit_transform(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size = 0.2,random_state=324)
X_train.shape[0] == y_train.shape[0]
```

```python
regressor = LinearRegression()
```

```python
regressor.get_params()
```

```python
regressor.fit(X_train,y_train)
```

```python
print("y intercept is :", regressor.intercept_)
print("coefficients are :", regressor.coef_)
```

```python
coeff_df = pd.DataFrame(regressor.coef_,factors,columns=['Coefficient'])
coeff_df
```

```python
list(zip(factors[0:8], regressor.coef_))
```

```python
y_predicted = regressor.predict(X_test)
```
